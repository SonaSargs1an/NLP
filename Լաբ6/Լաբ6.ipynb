```python
import re
import random
from collections import defaultdict

# === 1.
with open('robert_frost.txt', 'r', encoding='utf-8') as f:
    texts = f.read()
texts = texts.split('\n')

# === 2.
cleaned_texts = []
for text in texts:
    text = text.strip().lower()
    text = re.sub(r"[^a-z\s]", "", text)
    if text:
        cleaned_texts.append(text)

# === 3.
initial_counts = defaultdict(int)
for text in cleaned_texts:
    words = text.split()
    if words:
        first_word = words[0]
        initial_counts[first_word] += 1

total_texts = len(cleaned_texts)
initial_probabilities = {}
for word, count in initial_counts.items():
    initial_probabilities[word] = count / total_texts

# նորմալացում՝ հավանականությունների գումարը լինի 1
s_init = sum(initial_probabilities.values()) or 1
for w in list(initial_probabilities.keys()):
    initial_probabilities[w] = initial_probabilities[w] / s_init

# === 4. END
vocab = set()
for t in cleaned_texts:
    for w in t.split():
        vocab.add(w)
vocab.add("END")
V = len(vocab)

# === 5.
first_counts = defaultdict(lambda: defaultdict(int))
prev_counts = defaultdict(int)
second_counts = defaultdict(lambda: defaultdict(int))
pair_counts = defaultdict(int)

for text in cleaned_texts:
    words = text.split() + ["END"]

    # first-order
    for i in range(len(words) - 1):
        p = words[i]
        n = words[i + 1]
        first_counts[p][n] += 1
        prev_counts[p] += 1

    # second-order
    for i in range(len(words) - 2):
        p2 = words[i]
        p1 = words[i + 1]
        n = words[i + 2]
        second_counts[(p2, p1)][n] += 1
        pair_counts[(p2, p1)] += 1

# === 6.
P_first = {}
for prev, nxts in first_counts.items():
    total = prev_counts[prev]
    denom = total + V
    P_first[prev] = {}
    for w in vocab:
        cnt = nxts.get(w, 0)
        P_first[prev][w] = (cnt + 1) / denom

P_second = {}
for prev_pair, nxts in second_counts.items():
    total = pair_counts[prev_pair]
    denom = total + V
    P_second[prev_pair] = {}
    for w in vocab:
        cnt = nxts.get(w, 0)
        P_second[prev_pair][w] = (cnt + 1) / denom

# === 7.
LINES_TO_GENERATE = 4
deterministic = False

# === 8.
for _ in range(LINES_TO_GENERATE):
    r = random.random()
    cum = 0
    for w, p in initial_probabilities.items():
        cum += p
        if r <= cum:
            w1 = w
            break
    if w1 == "END":
        continue

    line = [w1]

    #first-order
    probs1 = P_first[w1] if w1 in P_first else initial_probabilities
    r = random.random()
    cum = 0
    for w, p in probs1.items():
        cum += p
        if r <= cum:
            w2 = w
            break
    if w2 == "END":
        print(" ".join(line))
        continue

    line.append(w2)

    #second-order
    while len(line) < 40:
        key = (line[-2], line[-1])
        if key in P_second:
            probs2 = P_second[key]
        elif line[-1] in P_first:
            probs2 = P_first[line[-1]]
        else:
            probs2 = initial_probabilities

        r = random.random()
        cum = 0
        for w, p in probs2.items():
            cum += p
            if r <= cum:
                nxt = w
                break
        if nxt == "END":
            break
        line.append(nxt)

    print(" ".join(line))
```
